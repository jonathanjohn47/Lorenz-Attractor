```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def lorenz(x,y,z, dt=0.01, n=10000):
    values = []
    for i in range(n):
        (a,b,c) = (28, 46.92, 4)
        dx = (a*(y-x))*dt
        dy = (x*(b - z) - y)*dt
        dz = (x*y - c*z)*dt

        x = x + dx
        y = y + dy
        z = z + dz
        
        values.append(x)
    
    return values

def meansquare(u,v):
    mean = None
    try:
        z = zip(u,v)
        sum = 0
        for i,j in z:
            sum = sum + (i-j)**2
        mean = sum/len(u)
    except:
        print("the two arrays differ in length")
    return mean

def averror(u,v):
    mean = None
    try:
        z = zip(u,v)
        sum = 0
        for i,j in z:
            sum = sum + np.absolute(i-j)
        mean = sum/len(u)
    except:
        print("the two arrays differ in length")
    return mean
```

```python
v = lorenz(1,1,1, dt = 0.01, n=5000)
#%matplotlib notebook
plt.plot(np.arange(len(v)), v, linewidth = 0.3)
u = lorenz(1.001,1,1, dt = 0.01, n=5000)
plt.plot(np.arange(len(u)), u, linewidth = 0.3)
plt.show()
print(np.sqrt(meansquare(u,v)))
print(averror(u,v))
```

```python
diff = []
z = zip(u,v)
for i,j in z:
    diff.append(i-j)
#%matplotlib notebook
plt.plot(np.arange(len(diff)), diff, linewidth = 0.4)
plt.show()
```

```python
def plotSpectrum(y,Fs):
    from numpy import sin, linspace, pi
    from pylab import plot, show, title, xlabel, ylabel, subplot
    from scipy import fft, arange
    """
    Plots a Single-Sided Amplitude Spectrum of y(t)
    """
    n = len(y) # length of the signal
    k = arange(n)
    T = n/Fs
    frq = k/T # two sides frequency range
    #frq = frq[range(n/2)] # one side frequency range

    Y = fft(y)/n # fft computing and normalization
    #Y = Y[range(n/2)]
    
    #%matplotlib notebook
    plt.plot(frq,abs(Y), linewidth = 0.2) # plotting the spectrum
    plt.xlabel('Freq (Hz)')
    plt.ylabel('|Y(freq)|')
```
